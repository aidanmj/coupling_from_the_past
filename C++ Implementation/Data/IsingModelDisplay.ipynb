```python
filename = '69895Ising_latt-100-2_45e-2.txt' #final sequence at 8.4 billion
SIZE = 100
```

```python
with open(filename) as f:
    line = f.readlines()[0]
    data = line.split(",")
    graphable = [[255,255,255] if z == "1" else [0,0,0] for z in data if z != ""]
```
